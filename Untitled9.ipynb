### (a) Problem Classification Decision Trees Address & Applications

A Classification Decision Tree is a model for predicting discrete or categorical outcomes, useful for problems where the goal is to assign an observation to one of several categories based on its features. The tree’s structure works by making a sequence of binary decisions (or splits) based on feature values, moving down the tree until it reaches a leaf node representing the predicted category.

Real-world applications include:
1. Medical Diagnosis: Classifying a patient as having a disease (or not) based on factors like symptoms, age, and test results.
2. Customer Segmentation: Categorizing customers into segments (e.g., “loyal” or “at-risk”) for personalized marketing.
3. Spam Detection: Classifying emails as “spam” or “not spam” based on keywords, frequency, and other features.
4. Loan Approval: Deciding whether an applicant is “approved” or “denied” based on credit score, income, and debt.

### (b) How Classification Decision Trees Make Predictions vs. Multiple Linear Regression

1. Classification Decision Tree:
- A Classification Decision Tree makes predictions by following a path from the root to a leaf node based on a sequence of rules applied to input features. Each node represents a question or rule about a feature (like "Is age < 30?"). Based on the answer, the model proceeds to the next node, and this continues until it reaches a leaf node, which indicates the predicted category.
- This process involves “interaction effects” since each split (or decision) can depend on prior splits, creating complex rules based on combinations of feature values.

2. Multiple Linear Regression:
- Multiple Linear Regression, on the other hand, is designed for predicting continuous values. It finds a linear relationship between input features and the target variable by calculating weights (coefficients) for each feature and combining them into a single formula: y = beta0 + beta1 * x1 + beta2 * x2 + ... + betan * xn   
- This linear form does not involve sequential decisions but rather treats each feature independently, estimating the target variable’s value directly through addition.

### 1. Accuracy  
- Scenario:Quality Control in Manufacturing
- Rationale: Accuracy is helpful when we want a general measure of correctness and there’s a relatively balanced dataset (similar number of defective and non-defective items). In quality control, it’s important to identify both correctly manufactured and defective products, so knowing the overall accuracy rate can indicate the effectiveness of the manufacturing process.

### 2. Sensitivity 
- Scenario: Medical Diagnostics for a Serious Disease (e.g., Cancer Screening)
- Rationale: Sensitivity is critical in cases where identifying all actual positives is paramount, even if some false positives occur. For serious diseases, missing a positive case (false negative) can have severe consequences. A high sensitivity ensures most, if not all, patients with the disease are identified, even if it means mistakenly flagging some healthy individuals.

### 3. Specificity
- Scenario: Drug Testing in Employment
- Rationale: Specificity is particularly important when the focus is on correctly identifying actual negatives and avoiding false positives. In drug testing, false positives could harm an employee’s career unjustly, so high specificity is desired to ensure that those who test positive truly have drugs in their system.

### 4. Precision  
- Scenario: Fraud Detection in Banking
- Rationale: Precision is useful in scenarios where we need to minimize false positives among detected positives, as false alarms can be costly or disruptive. In fraud detection, a high precision means that flagged transactions are more likely to be actual cases of fraud, reducing unnecessary investigations into legitimate transactions and ensuring resources are efficiently allocated.

Each metric offers a unique perspective based on the consequences of errors, allowing the selection of the most relevant metric for the scenario’s requirements.

## General Explanation of Predictions in clf2

The decision tree model clf2 makes predictions by evaluating the values of NumPages, Thick, and List Price for each book, following a sequence of decision rules from the root node to a leaf node. Here’s a simplified breakdown:

3. Leaf Nodes and Predictions: The model reaches a leaf node after a series of splits, where it makes the final classification.
   - At each leaf, the model assigns a prediction based on which class is more prevalent at that node.

```python
# Question 6
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
import numpy as np

# Define the feature sets for the test data
X_test_clf = ab_reduced_noNaN_test[['List Price']]
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']  # 'H' for hardcover

# Make predictions on the test set for both models
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Define a function to calculate sensitivity, specificity, and accuracy
def calculate_metrics(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    return np.round([sensitivity, specificity, accuracy], 3)

# Calculate metrics for clf
sensitivity_clf, specificity_clf, accuracy_clf = calculate_metrics(y_test, y_pred_clf)

# Calculate metrics for clf2
sensitivity_clf2, specificity_clf2, accuracy_clf2 = calculate_metrics(y_test, y_pred_clf2)

# Print the results
print("Model clf (List Price only):")
print(f"Sensitivity: {sensitivity_clf}")
print(f"Specificity: {specificity_clf}")
print(f"Accuracy: {accuracy_clf}\n")

print("Model clf2 (NumPages, Thick, and List Price):")
print(f"Sensitivity: {sensitivity_clf2}")
print(f"Specificity: {specificity_clf2}")
print(f"Accuracy: {accuracy_clf2}")
```
